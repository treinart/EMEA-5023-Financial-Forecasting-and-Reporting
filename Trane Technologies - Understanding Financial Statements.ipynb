<h2 style="color:#CFB87C;margin:0 0 14px 0;font-size:32px;">Financial Analysis Overview</h2>

<p style="font-size:17px;line-height:1.7;margin:0 0 20px 0;">
The financial statements point to a company that is performing well and has room to keep funding its operations. The most important finding is not just that Trane Technologies grew revenue. The better finding is that profit margins improved while the company continued to produce cash. That combination makes the 2025 results more convincing than top-line growth by itself.
</p>

<div style="background-color:#111111;border:1px solid #565A5C;border-top:6px solid #CFB87C;border-radius:10px;padding:18px;margin:16px 0;width:100%;box-sizing:border-box;">

<h3 style="color:#CFB87C;font-size:24px;margin:0 0 10px 0;">Profitability</h3>

<p style="font-size:17px;line-height:1.65;margin:0;color:#F8F8F8;">
Trane’s operating margin improved from 16.4% in 2023 to 18.6% in 2025. That matters because the company kept more profit from each dollar of revenue. Net margin was 13.7% in 2025, which shows that the improvement carried through beyond operating income and into earnings attributable to shareholders.
</p>

</div>

<div style="background-color:#111111;border:1px solid #565A5C;border-top:6px solid #CFB87C;border-radius:10px;padding:18px;margin:16px 0;width:100%;box-sizing:border-box;">

<h3 style="color:#CFB87C;font-size:24px;margin:0 0 10px 0;">Cash Generation</h3>

<p style="font-size:17px;line-height:1.65;margin:0;color:#F8F8F8;">
The cash flow statement supports the earnings result. Operating cash flow was higher than net earnings attributable to Trane, and company-adjusted non-GAAP free cash flow was about &#36;2.9B. That gives the company flexibility to reinvest, fund capital projects, support acquisitions, pay dividends, repurchase shares, and manage debt without relying only on new borrowing.
</p>

</div>
